# WeatherPy

```python
#Import dependencies
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import openweathermapy as owm
import requests
import time

#Add API key and Citipy
from config import owm_apikey
from citipy import citipy
from datetime import date

#Range of longitudes and latitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generating Cities List

```python
#Lists for cities and cooridnates
lat_lngs = []
cities = []

#Generates random coordinates
lat = np.random.uniform(low=-90.000, high=90.000, size=1500)
lng = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lat, lng)

#Identify nearest unique cities
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    if city not in cities:
        cities.append(city)
        
len(cities)
```

```python
#Split list of cities into lists of 50 max
n = 50 
final_cities = [cities[i * n:(i + 1) * n] for i in range((len(cities) + n - 1) // n )]  
```

## Performing API Calls

```python
#Settings dictionary
settings = {"units": "metric", "lang": "DE", "appid": owm_apikey}
```

```python
#Create empty list for all weather data
weather_data = []

#Set counters to 1
city_index = 1
list_index = 1

#Begin retrieval log
print("Beginning Data Retrieval")
print("---------------------------------")

#Nested loop loops through list of lists of cities and tries to add their info to weather_data list
#If try fails, city is dropped from list of cities and skip message appears
for city_group in final_cities:
    for city in city_group:  
        try:
            print(f"Processing Record {city_index} of Set {list_index} | {city}")
            weather_data.append(owm.get_current(city, **settings))
            city_index = city_index + 1
        except:
            print("City not found.  Skipping...")
            cities.remove(city)
            city_index = city_index + 1
    list_index = list_index + 1
    city_index = 1
    
print("---------------------------------")
print("Data Retrieval Complete")
print("---------------------------------")

#Day of retrieval
today = date.today().strftime("%m/%d/%Y")
```

## Converting API Data to Data Frame

```python
#List data to pull
summary = ["sys.country", "dt", "main.temp", "main.humidity", "wind.speed", "clouds.all", "coord.lat", "coord.lon"]

#Put data into dataframe
data = [response(*summary) for response in weather_data]
column_names = ["Country", "Date", "Temperature", "Humidity", "Wind Speed", "Cloudiness", "Latitude", "Longitude"]
weather_data = pd.DataFrame(data, index=cities, columns=column_names)

#Reformat city index as column
weather_data = weather_data.reset_index()
weather_data = weather_data.rename(columns={"index":"City"})

#Output to CSV and show header
weather_data.to_csv("weatherdata.csv", index=False, header=True)
weather_data.head()
```
